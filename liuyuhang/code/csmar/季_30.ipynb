```python
import numpy as np
import pandas as pd
```

```python
# 30. egr
# 季度频率。股本账面价值的季度变动百分比。

df1 = pd.read_csv('../../data/csmar/季_资产负债表.csv', usecols=['证券代码', '会计期间', '所有者权益合计'], low_memory=False)
df1['egr'] = (df1['所有者权益合计'] - df1.groupby('证券代码').shift(1)['所有者权益合计'])/df1.groupby('证券代码').shift(1)['所有者权益合计'].replace(0, np.nan)
df2 = df1[['证券代码', '会计期间', 'egr']]
df2.to_csv('../../output/csmar/季_30.csv', index=False, encoding='utf-8-sig')
df2
```
