```python
import numpy as np
import pandas as pd
```

```python
# 10. cashspr
# 季度频率。现金生产率，定义为季度末市值加上长期债务减去总资产除以现金和等价物。

df1 = pd.read_csv('../../data/csmar/季_相对价值指标.csv', usecols=['证券代码', '会计期间', '市值A'])
df2 = pd.read_csv('../../data/csmar/季_资产负债表.csv', usecols=['证券代码', '会计期间', '长期负债合计', '资产总计'], low_memory=False)
df3 = pd.read_csv('../../data/csmar/季_现金流量表(直接法).csv', usecols=['证券代码', '会计期间', '期末现金及现金等价物余额'], low_memory=False)
df4 = pd.merge(df2, df1, on=['证券代码', '会计期间'], how='left')
df5 = pd.merge(df4, df3, on=['证券代码', '会计期间'], how='left')
df5['cashspr'] = (df5['市值A'] + df5['长期负债合计'] - df5['资产总计'])/df5['期末现金及现金等价物余额'].replace(0, np.nan)
df6 = df5[['证券代码', '会计期间', 'cashspr']]
df6.to_csv('../../output/csmar/季_10.csv', index=False, encoding='utf-8-sig')
df6
```
