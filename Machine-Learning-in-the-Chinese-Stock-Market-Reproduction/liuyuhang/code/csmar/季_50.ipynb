```python
import pandas as pd
from utils.diff import diff
```

```python
# 50. orgcap
# 季度频率。资本化管理费用。该特征使用从CSMAR获得的费用数据，并根据Eisfeldt和Papanikolaou(2013)[5]的定义构建。

df1 = pd.read_csv('../../data/csmar/季_利润表.csv', usecols=['证券代码', '会计期间', '管理费用', '销售费用'])
df1 = diff(df1, col=['管理费用', '销售费用'])
df1['orgcap'] = df1['管理费用'] + df1['销售费用']
df2 = df1[['证券代码', '会计期间', 'orgcap']]
df2.to_csv('../../output/csmar/季_50.csv', index=False, encoding='utf-8-sig')
df2
```
