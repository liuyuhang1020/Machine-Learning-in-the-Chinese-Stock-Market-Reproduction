```python
import numpy as np
import pandas as pd
from utils.diff import diff
```

```python
# 74.  salecash
# 季度频率。季度销售额除以现金和现金等价物。

df1 = pd.read_csv('../../data/csmar/季_资产负债表.csv', usecols=['证券代码', '会计期间', '货币资金'], low_memory=False)
df2 = pd.read_csv('../../data/csmar/季_利润表.csv', usecols=['证券代码', '会计期间', '营业收入'])
df3 = pd.merge(df1, df2, on=['证券代码', '会计期间'], how='left')
df3 = diff(df3, col=['营业收入'])
df3['salecash'] = df3['营业收入']/df3['货币资金'].replace(0, np.nan)
df4 = df3[['证券代码', '会计期间', 'salecash']]
df4.to_csv('../../output/csmar/季_74.csv', index=False, encoding='utf-8-sig')
df4
```
