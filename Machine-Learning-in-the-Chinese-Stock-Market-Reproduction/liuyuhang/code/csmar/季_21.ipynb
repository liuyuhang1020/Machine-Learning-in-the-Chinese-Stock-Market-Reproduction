```python
import numpy as np
import pandas as pd
```

```python
# 21.  chtx
# 季度频率。从第t-1季度到第t季度，税收的百分比变化。

df1 = pd.read_csv('../../data/csmar/季_资产负债表.csv', usecols=['证券代码', '会计期间', '应交税费'], low_memory=False)
df1['chtx'] = (df1['应交税费'] - df1.groupby('证券代码').shift(1)['应交税费'])/df1.groupby('证券代码').shift(1)['应交税费'].replace(0, np.nan)
df2 = df1[['证券代码', '会计期间', 'chtx']]
df2.to_csv('../../output/csmar/季_21.csv', index=False, encoding='utf-8-sig')
df2
```
