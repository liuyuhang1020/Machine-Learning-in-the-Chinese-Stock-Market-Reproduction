```python
# 上市公司公告
for ts_code in tqdm(ts_code_list):
    
    last = end
    df1_list = []
    while True:
        while True:
            try:
                df1 = pro.query('anns', 
                            ts_code=ts_code,
                            start_date=start,
                            end_date=last,
                            fields='ts_code, ann_date, ann_type, title, pub_time' 
                            )
                break
            except:
                sleep(1)
                continue
        df1 = df1.rename(columns=
                    {'ts_code': 'TS股票代码', 'ann_date': '公告日期', 'ann_type': '公告类型', 'title': '公告标题', 'pub_time': '公告发布时间'}
                    )
        df1 = df1.sort_values(by=['公告日期'])
        if not list(df1['公告日期']):
            break
        elif last == list(df1['公告日期'])[0]:
            break
        else:
            last = list(df1['公告日期'])[0]
            df1_list.append(df1)

    if df1_list:
        df = pd.concat(df1_list, axis=0)
        df = df.drop_duplicates(subset=['公告日期'])
        df.to_csv('../../data/tushare/日_上市公司公告.csv', index=False, encoding='utf-8-sig', mode='a', header=False)
    else:
        continue

result = pd.read_csv('../../data/tushare/日_上市公司公告.csv')
result = result.sort_values(by=['TS股票代码', '公告日期']).reset_index().drop(columns=['index'])
result = result.drop_duplicates()
skt_df = pd.read_csv('../../data/tushare/股票列表.csv')[['TS股票代码', '证券代码']]
result = pd.merge(result, skt_df, on=['TS股票代码'], how='left')
result.to_csv('../../data/tushare/日_上市公司公告.csv', index=False, encoding='utf-8-sig')
result
```

```python
# 每日指标
for ts_code in tqdm(ts_code_list):
    
    last = end
    df1_list = []
    while True:
        while True:
            try:
                df1 = pro.query('daily_basic', 
                            ts_code=ts_code,
                            start_date=start,
                            end_date=last,
                            fields='ts_code, trade_date, close, turnover_rate, turnover_rate_f, volume_ratio, pe, pe_ttm, pb, ps, ps_ttm, dv_ratio, dv_ttm, total_share, float_share, free_share, total_mv, circ_mv' 
                            )
                break
            except:
                sleep(1)
                continue
        df1 = df1.rename(columns=
                    {'ts_code': 'TS股票代码', 'trade_date': '交易日期', 'close': '收盘价', 'turnover_rate': '换手率', 'turnover_rate_f': '换手率(自由流通股)', 'volume_ratio': '量比', 'pe': '市盈率', 'pe_ttm': '市盈率(TTM)', 'pb': '市净率', 'ps': '市销率', 'ps_ttm': '市销率(TTM)', 'dv_ratio': '股息率', 'dv_ttm': '股息率(TTM)', 'total_share': '总股本', 'float_share': '流通股本', 'free_share': '自由流通股本', 'total_mv': '总市值', 'circ_mv': '流通市值'}
                    )
        df1['换手率'] = df1['换手率']/100
        df1['换手率(自由流通股)'] = df1['换手率(自由流通股)']/100
        df1['股息率'] = df1['股息率']/100
        df1['股息率(TTM)'] = df1['股息率(TTM)']/100
        df1['总股本'] = 10000*df1['总股本']
        df1['流通股本'] = 10000*df1['流通股本']
        df1['自由流通股本'] = 10000*df1['自由流通股本']
        df1['总市值'] = 10000*df1['总市值']
        df1['流通市值'] = 10000*df1['流通市值']
        df1 = df1.sort_values(by=['交易日期'])
        if not list(df1['交易日期']):
            break
        elif last == list(df1['交易日期'])[0]:
            break
        else:
            last = list(df1['交易日期'])[0]
            df1_list.append(df1)

    if df1_list:
        df = pd.concat(df1_list, axis=0)
        df = df.drop_duplicates(subset=['交易日期'])
        df.to_csv('../../data/tushare/日_每日指标.csv', index=False, encoding='utf-8-sig', mode='a', header=False)
    else:
        continue

result = pd.read_csv('../../data/tushare/日_每日指标.csv')
result = result.sort_values(by=['TS股票代码', '交易日期']).reset_index().drop(columns=['index'])
result = result.drop_duplicates(subset=['TS股票代码', '交易日期'])
skt_df = pd.read_csv('../../data/tushare/股票列表.csv')[['TS股票代码', '证券代码']]
result = pd.merge(result, skt_df, on=['TS股票代码'], how='left')
result.to_csv('../../data/tushare/日_每日指标.csv', index=False, encoding='utf-8-sig')
result
```
